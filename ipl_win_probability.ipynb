```python
#MODEL USED = LOGISTIC REGRESSION

import pandas as pd
import numpy as np
```

```python
match=pd.read_csv('/content/matches.csv')
delivery=pd.read_csv('/content/deliveries.csv')
```

```python
match.head()
```

```python
delivery.head()
```

```python
total_score_df=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df
```

```python
#remove the 2nd innings runs from dataset
total_score_df=total_score_df[total_score_df['inning']==1]
```

```python
total_score_df
```

```python
#merge the dataset ==>  total_score_df and match dataset
match_df = match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df.head()
```

```python
match_df['team1'].unique()
```

```python
teams=[
    'Sunrisers Hyderabad',
    'Mumbai Indians',
    'Royal Challengers Bangalore',
    'Kolkata Knight Riders',
    'Kings XI Punjab',
    'Chennai Super Kings',
    'Rajasthan Royals',
    'Delhi Capitals'

]
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils','Delhi Capitals')


match_df['team1']=match_df['team1'].str.replace('Deccan Charges','Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Charges','Sunrisers Hyderabad')
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
match_df
```

```python
##delete those row in which rain is occured

match_df=match_df[match_df['dl_applied']==0]
```

```python
match_df['dl_applied'].value_counts()
```

```python
match_df=match_df[['match_id','city','winner','total_runs']]
match_df
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
```

```python
# SELECT ONLY 2ND INNING ROWS

delivery_df=delivery_df[delivery_df['inning']==2]
```

```python
delivery_df
```

```python
delivery_df['total_runs_y'].cumsum()
```

```python
delivery_df['current_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df
```

```python
#runs left column
delivery_df['runs_left']=  delivery_df['total_runs_x'] - delivery_df['current_score']
```

```python
delivery_df
```

```python
#balls left column
delivery_df['balls_left']= 126 - (delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df
```

```python
#calculating wickets_left column

delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna('0')
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x=='0' else '1')
delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype('int')

#how much wickets are there after each ball
wickets=delivery_df.groupby('match_id')['player_dismissed'].cumsum().values

#wicket_left column
delivery_df['wickets']=10-wickets
```

```python
delivery_df
```

```python
#calculating current_run_rate ==> crr=runs/overs

delivery_df['crr'] = (delivery_df['current_score']*6)/(120 - delivery_df['balls_left'])
```

```python
#calculating required_run_rate ==> crr=runs_left/over_left

delivery_df['rrr'] = (delivery_df['runs_left']*6)/delivery_df['balls_left']
```

```python
delivery_df
```

```python
#making winner column into 0 and 1

def result(row):

  return 1 if row['batting_team'] == row['winner'] else 0
```

```python
delivery_df['result'] = delivery_df.apply(result,axis=1)
```

```python
delivery_df
```

```python
final_df = delivery_df[['batting_team','bowling_team','city','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df.isnull().sum()
```

```python
final_df.dropna(inplace=True)
```

```python
final_df = final_df[final_df['balls_left'] != 0]
```

```python
#train test split

X=final_df.iloc[:,:-1]
y=final_df.iloc[:,-1]

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse=False,drop='first'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',LogisticRegression(solver='liblinear'))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
pipe.predict_proba(X_test)[0]
```
